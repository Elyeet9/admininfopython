## Datos

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Importe matplotlib.pyplot como plt y configure % matplotlib inline si est√° utilizando el cuaderno jupyter. ¬øQu√© comando usa si no est√° usando el cuaderno jupyter?**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Ejercicio 1

**Siga estos pasos:**

* ** Cree un objeto de figura llamado fig usando plt.figure () **
* ** Use add_axes para agregar un eje al lienzo de la figura en [0,0,1,1]. Llame a este nuevo eje ax. **
* ** Grafique (x, y) en esos ejes y establezca las etiquetas y t√≠tulos para que coincidan con el gr√°fico a continuaci√≥n: **

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Titulo (no me acuerdo üò¢)')
```

## Ejercicio 2
** Crea un objeto de figura y ponle dos ejes, ax1 y ax2. Ubicado en [0,0,1,1] y [0.2,0.5,.4,.4] respectivamente. **

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

** Ahora grafique (x, y) en ambos ejes. Y llama a tu objeto figura para mostrarlo. **

```python
ax1.plot(x,y)
ax2.plot(x,y)
fig
```

## Ejercicio 3

** Cree la gr√°fica de abajo agregando dos ejes a un objeto figura en [0,0,1,1] y [0.2,0.5, .4, .4] **

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.3,0.5,.4,.4])
```

** Ahora use las matrices **x**, **y** y **z** para recrear el diagrama a continuaci√≥n. Observe los xlimits y los l√≠mites y en la gr√°fica insertada: **

```python
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_xlim(0,100)
ax1.set_ylim(0,200)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20.0,22.0)
ax2.set_ylim(30,50)
ax1.plot(x,y)
ax2.plot(x,y)
fig
```

## Ejercicio 4

** Utilice plt.subplots(nrows = 1, ncols = 2) para crear el gr√°fico a continuaci√≥n. **

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
```

** Ahora grafique (x,y) y (x,z) en los ejes. Juega con el ancho de l√≠nea y el estilo **

```python
axes[0].plot(x,y,color = 'blue', ls = '--', lw = 2)
axes[1].plot(x,z,'r')
fig
```
